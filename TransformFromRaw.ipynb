```
def sub_unmount(str_path):
    if any(mount.mountPoint == str_path for mount in dbutils.fs.mounts()):
        dbutils.fs.unmount(str_path)

sub_unmount("/mnt/raw")
```

```
input_path='/mnt/raw/salary.csv'
```

```
df=spark.read.format("csv").load(input_path)
df.toPandas()
```

```
#!st Column as Header
x= sc.textFile(input_path).zipWithIndex().filter(lambda a:a[1]>=0).map(lambda a:a[0].split(","))
x.collect()
```

```
column_rdd=x.first()
```

```
column_rdd
```

```
#Remove Column Row
x=x.filter(lambda a: a!= column_rdd)

# Combine the data and Column
data=x.toDF(column_rdd)
```

```
data.toPandas()
```
